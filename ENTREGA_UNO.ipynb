# Unión de datos

```python
df = df_employee_survey.merge(df_general, on="EmployeeID")
df = df.merge(df_manager_survey, on="EmployeeID")
df = df.merge(df_time_work, on="EmployeeID")

df.info()
df.head()
```

```python
# Obtenemos las filas con todos los datos vacíos
df_empty = df[df.isnull().all(axis=1)]
df_empty
```

```python
# Obtenemos el número de datos vacíos por columna
df.isnull().sum()
```

```python
# Solo tipo objeto
# Obtenemos las columnas con datos vacíos
df_null = df.isnull().sum()

# Imprimimos las columnas con datos vacíos
print(df_null[df_null > 0].index)
```

```python
# todas las variables, objetos, float, enteros
# Obtenemos las columnas con datos vacíos
df_null = df.isnull().any()

# Imprimimos las columnas con datos vacíos
print(df_null[df_null == True].index)
```
